```python
pd.set_option("display.max_columns", 100)
url = "https://raw.githubusercontent.com/cs6220/cs6220.spring2019/master/data/AmesHousing.txt"
df2 = pd.read_csv(url, sep="\t")
```

```python
df_original=df2
df2=df2.drop(['Order','PID','SalePrice'],axis=1)
df2
```

### 2.1 Imputation

Âˆ - [5 pts] How many features have missing values?

```python
values_missing = pd.isnull(df2).sum()
print(values_missing)
```

```python
count = 0
for val in values_missing:
    if val > 0:
        count+=1
print("Number of features missing values:", count)
```


Âˆ - [5 pts] Fill each missing nominal feature value with the string â€œMissingâ€.


```python
df2.fillna("Missing")
```


Âˆ - [5 pts] Interpolate each missing numeric feature value using linear interpolation.

```python
numeric_cols = [col for col in df2.columns if pd.api.types.is_numeric_dtype(df2[col])]
for col in numeric_cols:
    df2[[col]]=df2[[col]].interpolate(method='linear')
df2=df2.fillna("Missing")
df2
```

### 2.2 Standardization

```python
from sklearn.preprocessing import scale
df2_scaled = df2.copy()
df2_scaled[numeric_cols] = scale(df2_scaled[numeric_cols])
df2_scaled[numeric_cols]
```

Box plot before standarlization

```python
plt.figure(figsize=(12, 8)) 
box_before = df2.boxplot(grid=False,return_type='axes')
plt.xticks(rotation=45)
plt.show()
```

Box plot after standarlization

```python
plt.figure(figsize=(12, 8)) 
box_after = df2_scaled.boxplot(grid=False,return_type='axes')
plt.xticks(rotation=45)
plt.show()
```

Compared with two box plots, we can see stardadized one makes more sense for that the range is smaller and the distribution of each feature is more readable.

Before standadization, the outlier feature is Lot Area.

After standadization, the outlier feature is Misc Val.


### 2.3 Feature Selection

- [10 pts] To get an idea of their relative importance, estimate the mutual information between the numeric features and the class column, â€˜SalePriceâ€™. 

```python
from sklearn.datasets import make_regression
from sklearn.feature_selection import mutual_info_regression
```

```python
mutual_info = mutual_info_regression(df2_scaled[numeric_cols], df_original['SalePrice'])
mutual_info
```

```python
mi_results = pd.DataFrame({'Feature': df2_scaled[numeric_cols].columns, 'MI score': mutual_info})
mi_results = mi_results.sort_values(by='MI score', ascending=False)
```

```python
mi_results
```

- [10 pts] What are the top 5 numeric features ranked by mutual information? Note that features with a higher estimated mutual information are considered more informative.

```python
mi_results.head()
# top 5 numeric features: Overall Qual, Gr Liv Area, Garage Area, Year Built, Total Bsmt SF
```
