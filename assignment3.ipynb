```python
import pandas as pd 
pd.set_option("max_colwidth", 150)

f = "https://github.com/cs6220/cs6220.spring2019/raw/master/data/Online%20Retail.xlsx"
df = pd.read_excel(f)
basket = (df[df["Country"] == "United Kingdom"].groupby(["InvoiceNo", "Description"])["Quantity"].sum().unstack().reset_index().fillna(0).set_index("InvoiceNo")) # transform transactions into baskets of items
basket_sets = basket.map(lambda x: True if x >=1 else False) # convert counts to booleans
```

## 1.1 Frequent Itemset Generation

**1. What are the top 5 1-itemsets with the highest support?**

```python
basket_sets.head()
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
freq_items=apriori(basket_sets, min_support=0.02, use_colnames=True)
freq_items.sort_values(by='support',ascending=False)
```

```python
freq_1itemset = freq_items[freq_items['itemsets'].apply(lambda x: len(x) == 1)]
```

```python
top5_freq_1itemset = freq_1itemset.sort_values(by='support',ascending=False).head()
top5_freq_1itemset
```

Hence, top 5 itemset with highest support are: 
1. WHITE HANGING HEART T-LIGHT HOLDER
2. JUMBO BAG RED RETROSPO
3. REGENCY CAKESTAND 3 TIER
4. PARTY BUNTING
5. LUNCH BAG RED RETROSPOT

**2. What are the top 5 2-itemsets with the highest support?**

```python
freq_2itemsets = freq_items[freq_items['itemsets'].apply(lambda x: len(x) == 2)]
```

```python
top5_freq_2itemsets = freq_2itemsets.sort_values(by='support',ascending=False).head()
top5_freq_2itemsets
```

**3. What is the highest support value for the 1-itemsets?**

From top5_freq_1itemset above, we can see the highest support value for the 1-itemsets is **0.098276**	

**4. What is the highest support value for the 2-itemsets?**

From top5_freq_2itemsets above, we can see the highest support value for the 2-itemsets is **0.035617**
